```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
ax[0,0].scatter(batters["avg"],batters["strike_rate"],color="red")
ax[0,1].scatter(batters["avg"],batters["runs"])
ax[1,0].hist(batters["avg"])
ax[1,1].hist(batters["runs"])

plt.show()
```

```python
fig=plt.figure()

ax1=fig.add_subplot(2,2,1)# here 3 parameters,.1st tells us how many rows,2nd specifies how many cols and 3rd specify nth plot
ax1.scatter(batters["avg"],batters["strike_rate"],color="red")

ax2=fig.add_subplot(2,2,2)
ax2.hist(batters["avg"])

ax3=fig.add_subplot(2,2,3)
ax3.hist(batters["runs"])
```

**3D surface plot**

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
z=xx**2+yy**2
z.shape
```

```python
fig=plt.figure(figsize=(12,8))

ax=plt.subplot(projection="3d")

p=ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```

```python
z=np.sin(xx)+np.cos(yy)
fig=plt.figure(figsize=(12,8))

ax=plt.subplot(projection="3d")

p=ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(p)
```
